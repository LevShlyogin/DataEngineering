# Первая задача

```python
import numpy as np
import json

# Загрузка матрицы из файла с форматом npy
matrix = np.load("/content/matrix_54.npy")

# Подсчет суммы всех элементов
sum_all = np.sum(matrix)

# Подсчет среднего арифметического всех элементов
avg_all = np.mean(matrix)

# Подсчет суммы и среднего арифметического главной и побочной диагоналей
sum_main_diagonal = np.sum(np.diag(matrix))
avg_main_diagonal = np.mean(np.diag(matrix))

sum_secondary_diagonal = np.sum(np.diag(np.fliplr(matrix)))
avg_secondary_diagonal = np.mean(np.diag(np.fliplr(matrix)))

# Нахождение максимального и минимального значений
max_value = np.max(matrix)
min_value = np.min(matrix)

# Запись полученных значений в json
result = {
    'sum': float(sum_all),
    'avr': float(avg_all),
    'sumMD': float(sum_main_diagonal),
    'avrMD': float(avg_main_diagonal),
    'sumSD': float(sum_secondary_diagonal),
    'avrSD': float(avg_secondary_diagonal),
    'max': float(max_value),
    'min': float(min_value)
}

with open('result.json', 'w') as file:
    json.dump(result, file)

# Нормализация и сохранение матрицы в формате npy
normalized_matrix = (matrix - np.min(matrix)) / (np.max(matrix) - np.min(matrix))
np.save("normalized_matrix.npy", normalized_matrix)
```

# Вторая задача

```python
import numpy as np
import os

# Загрузка матрицы из файла .npy
matrix = np.load('/content/matrix_54_2.npy')

# Создание пустых массивов x, y, z
x = []
y = []
z = []

# Отбор значений, превышающих 554, и разнесение индексов в массивы x и y
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        if matrix[i, j] > 554:
            x.append(i)
            y.append(j)
            z.append(matrix[i, j])

# Сохранение полученных массивов в файл формата npz
np.savez('result.npz', x=x, y=y, z=z)

# Сохранение файла с использованием метода np.savez_compressed()
np.savez_compressed('result_compressed.npz', x=x, y=y, z=z)

# Сравнение размеров полученных файлов
original_size = os.path.getsize('/content/result.npz')
compressed_size = os.path.getsize('/content/result_compressed.npz')

print("Размер исходного файла:", original_size, "bytes")
print("Размер сжатого файла:", compressed_size, "bytes")
```

# Третья задача

```python
import json
import msgpack
import os

# Считываем массив объектов из файла JSON
with open("/content/products_54.json", "r") as json_file:
    data = json.load(json_file)

# Собираем агрегированную информацию
    products = dict()

    for item in data:
        if item['name'] in products:
            products[item['name']].append(item['price'])
        else:
             products[item['name']] = list()
             products[item['name']].append(item['price'])

    result = list()

    for name, prices in products.items():
        sum_price = 0
        max_price = prices[0]
        min_price = prices[0]
        size = len(prices)
        for price in prices:
            sum_price += price
            max_prise = max(max_price, price)
            min_price = min(min_price, price)
        result.append({
            'name': name,
            'max': max_price,
            'min': min_price,
            'avr': sum_price/size
        })

# Сохраняем информацию в формате JSON
with open("aggregated_info.json", "w") as json_output_file:
    json_output_file.write(json.dumps(result))

# Сохраняем информацию в формате MessagePack
with open("aggregated_info.msgpack", "wb") as msgpack_output_file:
    msgpack_output_file.write(msgpack.dumps(result))

# Показываем размеры файлов
json_file_size = os.path.getsize("aggregated_info.json")
msgpack_file_size = os.path.getsize("aggregated_info.msgpack")

print("Размер файла JSON:", json_file_size, "байт")
print("Размер файла MessagePack:", msgpack_file_size, "байт")
```

# Четвёртая задача

```python
import json
import pickle

# Считываем данные о товарах из файла формата pkl
with open("/content/products_54.pkl", "rb") as f:
    products = pickle.load(f)

# Считываем данные о товарах из файла формата json
with open("/content/price_info_54.json") as f:
    price_info = json.load(f)

# Функция для обновления цен товаров
def update_price(product, price_info):
    method = price_info["method"]
    if method == "sum":
        product["price"] += price_info["param"]
    elif method == "sub":
        product["price"] -= price_info["param"]
    elif method == "percent+":
        product["price"] *= (1 + price_info["param"])
    elif method == "percent-":
        product["price"] *= (1 - price_info["param"])
    product["price"] = round(product["price"], 2)

# Разложим в карту информацию об обновлении цен
price_info_dict = dict()
for item in price_info:
    price_info_dict[item["name"]] = item

# Вызываем фуцнкцию
for product in products:
    current_price_info = price_info_dict[product["name"]]
    update_price(product, current_price_info)

# Сохраняем информацию в формате pkl
with open("products_updated.pkl", "wb") as f:
    f.write(pickle.dumps(products))
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"/content/master.csv", encoding = "utf-8")
df
```

```python
numeric_columns = df.select_dtypes(include='number').columns
numeric_stats = df[numeric_columns].describe().to_dict()
numeric_stats
```

```python
df['country'] = df['country'].astype('category').cat.codes
df['sex'] = df['sex'].astype('category').cat.codes
df['age'] = df['age'].astype('category').cat.codes
df['country-year'] = df['country-year'].astype('category').cat.codes
df['generation'] = df['generation'].astype('category').cat.codes
df
```

```python
p1 = df['country'].value_counts().to_dict()
p2 = df['sex'].value_counts().to_dict()
p3 = df['age'].value_counts().to_dict()
p4 = df['country-year'].value_counts().to_dict()
p5 = df['generation'].value_counts().to_dict()
```

```python
result = numeric_stats | p1 | p2 | p3 | p4 | p5
result
```

```python
# Сохраняем результаты в формате JSON
with open("rs.json", "w") as json_output_file:
    json_output_file.write(json.dumps(result))
```
